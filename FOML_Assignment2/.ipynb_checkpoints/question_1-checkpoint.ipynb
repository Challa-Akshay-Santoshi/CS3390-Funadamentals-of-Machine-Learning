```python
# QUESTION 1: Principal Component Analysis

import numpy as np
from matplotlib import pyplot as plt
from scipy.io import loadmat, savemat
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Collecting data from train_data and test_32x32

np.random.seed(20)
train_data = loadmat('train_data.mat')
test_data = loadmat('test_32x32.mat')
train_images = np.array(train_data['X'])
test_images = np.array(test_data['X'])
height, width, channels, tr_num_samples = train_images.shape
test_num_samples = test_images.shape[3]

train_labels = train_data['y']
test_labels = test_data['y']
test_labels[test_labels == 10] = 0

train_images = np.moveaxis(train_images, -1,0)
test_images = np.moveaxis(test_images, -1,0)
train_images = train_images.reshape(tr_num_samples,-1) / 255.0
test_images = test_images.reshape(test_num_samples,-1) / 255.0
```

```python
print(train_images.shape)
print(train_labels.shape)
print(test_images.shape)
print(test_labels.shape)
print(tr_num_samples)
print(test_num_samples)
```

```python
# QUESTION 1(a)

# Perform PCA on SVHN data set. 
# Find how many top eigenvectors are required to keep the proportion of variance above 0.9. 
# Plot PoV against number of eigenvectors.
```

```python
# Standardize the data

scaler = StandardScaler()
train_std = scaler.fit_transform(train_images)

# Perform PCA

pca = PCA()
pca.fit(train_std)
```

```python
# Calculate the cumulative variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

# To get the number of top eigenvectors required to keep poV above 0.9
number = np.argmax(cumulative_variance >= 0.9) + 1

print(f'Number of top eigenvectors required to keep the poV above 0.9 is {number}')
```

```python
# Plot PoV against number of eigenvectors
x_int = range(1, len(cumulative_variance)+1)
plt.plot(x_int, cumulative_variance, marker='o')

plt.title('Proportion of Variance (PoV) vs. Number of Eigenvectors')
plt.xlabel('Number of eigen vectors')
plt.ylabel('Proportion of Variance')
plt.grid(True)
plt.show()
```

```python
# Limiting the plot to get better view of the top eigen vectors
x_int = range(1, len(cumulative_variance)+1)
plt.plot(x_int, cumulative_variance, marker='o')

plt.title('Proportion of Variance (PoV) vs. Number of Eigenvectors limited to poV <= 0.9')
plt.xlabel('Number of eigen vectors')
plt.ylabel('Proportion of Variance')
plt.grid(True)
plt.xlim(1,number,1)
plt.show()
```

```python
# QUESTION 1(b)

# Visualize top 10 eigenvectors and provide reconstruction of 10 SVHN samples (one from each class) using top 10 eigenvectors.
```

```python
# Extract top 10 eigenvectors
top_eigenvectors = pca.components_[:10, :]
print(top_eigenvectors)
# Visualize top 10 eigenvectors
top_eigenvalues = pca.explained_variance_[:10]
# Create a figure
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot eigenvectors as bars
ax1.bar(range(1, len(top_eigenvalues) + 1), top_eigenvalues, alpha=0.7, align='center', label='Eigenvalues')
ax1.set_xlabel('Principal Components')
ax1.set_ylabel('Eigenvalues', color='tab:blue')
ax1.tick_params('y', colors='tab:blue')
ax1.set_title('Top 10 Eigenvectors and Eigenvalues')
```

```python
# Projection-matrix W
W = top_eigenvectors.T

fig, axes = plt.subplots(2, 5, figsize=(12, 6))
for digit in range(10):
    class_index = np.where(train_labels==digit)[0][16]

    class_sample = train_std[class_index]
    new_sample = np.dot(np.dot(class_sample, W), W.T) + scaler.mean_
    # Reshape for plotting
    original_sample = class_sample.reshape(height, width, channels)
    reconstructed_sample = new_sample.reshape(height, width,channels)
    
    # Plot original and reconstructed samples
    axes[digit // 5, digit % 5].imshow(np.hstack([original_sample, reconstructed_sample]), cmap='gray')
    axes[digit // 5, digit % 5].set_title(f'Digit {digit}')

plt.suptitle('Original vs Reconstructed Samples using Top 10 Eigenvectors')
plt.show()
```

```python
# QUESTION 1(c)

# Run k-NN (for k=5 and k=7) on raw data and data obtained after PCA dimensionality
# reduction for dimension as found in part (a) and for dimension 10 as in part (b). 
# Provide the accuracy of the predictions on the test data set for these various cases and discuss your observations.
```

```python
# Get data required in kNN and to predict accuracy scores

test_std = scaler.transform(test_images)

# Raw data
train_raw = train_images * 255.0
test_raw = test_images * 255.0

# Data to perform pca with components obtained from part(a)
pca1 = PCA(n_components = number)
train_pca1 = pca1.fit_transform(train_std)
test_pca1 = pca1.transform(test_std)

# Data to perform pca with components as 10
pca2 = PCA(n_components = 10)
train_pca2 = pca2.fit_transform(train_std)
test_pca2 = pca2.transform(test_std)
```

```python
# Performing k-NN for k=5 and k=7
k_list = [5,7]

for k in k_list:

    # k-NN train on raw data and accuracy of the test data
    knn_raw = KNeighborsClassifier(n_neighbors=k)
    knn_raw.fit(train_raw, train_labels.flatten())
    predictions = knn_raw.predict(test_raw)
    accuracy = accuracy_score(test_labels.flatten(), predictions)
    print(f'Accuracy of the predictions on raw test data for k={k}: {accuracy:.5f}')

    # k-NN train on PCA for dimensions reduced to 'number' (27)
    knn_pca1 = KNeighborsClassifier(n_neighbors=k)
    knn_pca1.fit(train_pca1, train_labels.flatten())
    predictions = knn_pca1.predict(test_pca1)
    accuracy = accuracy_score(test_labels.flatten(), predictions)
    print(f'Accuracy of the predictions on test data (PCA for dimensions = 27) for k={k}: {accuracy:.5f}')

    # k-NN train on PCA for dimensions reduced to 10
    knn_pca2 = KNeighborsClassifier(n_neighbors=k)
    knn_pca2.fit(train_pca2, train_labels.flatten())
    predictions = knn_pca2.predict(test_pca2)
    accuracy = accuracy_score(test_labels.flatten(), predictions)
    print(f'Accuracy of the predictions on test data (PCA for dimensions = 10) for k={k}: {accuracy:.5f}')
```

```python
## Observations

# Data with PCA performed on high dimension as found in part(a) retains 90% of the information
# but data with PCA performed on dimension equal to 10 will result in loss of information.
# Therefore we can see that accuracy of the predictions on raw data is greater than that on PCA limited to dimension = 10.
# In high-dimensional spaces, the distance between data points tends to increase and this can sometimes negatively impact 
# k-NN performance. By reducing dimensionality through PCA, it can lead to higher accuracy.
# We can also observe that with the increase in the value of k, the accuracy has increased.
```
