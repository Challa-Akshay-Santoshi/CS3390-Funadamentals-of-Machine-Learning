```python
# Introducing feature and target variables
feature = data.drop(['time_taken(mins)'], axis=1)
target = data['time_taken(mins)']
X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size=0.2, random_state=42)
X_train.shape, X_test.shape
```

IMPLEMENTATION OF DECISION TREE CLASSIFIER

```python
# Create a Decision Tree classifier
clf = DecisionTreeClassifier()

# Train the classifier on the training data
# print(X_train, y_train)
clf.fit(X_train, y_train)

# Use the trained model to make predictions on the test data
y_pred = clf.predict(X_test)
```

IMPLEMENTATION OF LINEAR REGRESSION

```python
lin_model = LinearRegression()
lin_model.fit(X_train, y_train)
lin_ypred = lin_model.predict(X_test)
```

IMPLEMENTATION OF GRADIENT BOOSTING REGRESSION

```python
# Create a Gradient Boosting Regressor with specified hyperparameters
gb_model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)

# Train the regressor on the training data
gb_model.fit(X_train, y_train)

# Use the trained model to make predictions on the test data
gb_ypred = gb_model.predict(X_test)
```

IMPLEMENTATIOIN OF K-NEAREST REGRESSION

```python
# Create a K-nearest Neighbor Regressor with a specified number of neighbors (K)
kn_model = KNeighborsRegressor(n_neighbors=5)

# Train the regressor on the training data
kn_model.fit(X_train, y_train)

# Use the trained model to make predictions on the test data
kn_ypred = kn_model.predict(X_test)
```

```python
print('Root Mean Squared Error for decision forest tree:',np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('Root Mean Squared Error for linear regression:',np.sqrt(metrics.mean_squared_error(y_test, lin_ypred)))
print('Root Mean Squared Error for gradient boosting regression:',np.sqrt(metrics.mean_squared_error(y_test, gb_ypred)))
print('Root Mean Squared Error for K-Nearest Model regression:',np.sqrt(metrics.mean_squared_error(y_test, kn_ypred)))
```

```python
print('R-Squared value for decision forest tree:',metrics.r2_score(y_test, y_pred))
print('R-Squared value for linear regression:',metrics.r2_score(y_test, lin_ypred))
print('R-Squared value for gradient boosting regression:',metrics.r2_score(y_test, gb_ypred))
print('R-Squared value for K-Nearest Model regression:',metrics.r2_score(y_test, kn_ypred))
```

CONCLUSION:

    As we see, according to the evaluation metrics we have considered, the linear regression model is not best fit. Among the four models we have considered, Gradient Boosting Regression is fitting best. If we see the R-squared values, except for gradient boosting method, remaining are negative values, which indicates that they are poor fits for our data. Only gradient boosting regression is best fit for our data
